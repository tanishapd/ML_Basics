```python
percentile25=df['Age'].quantile(0.25)
percentile75=df['Age'].quantile(0.75)
```

```python
percentile25
```

```python
percentile75
```

```python
iqr=percentile75-percentile25
print(iqr)
```

```python
upper_limit=percentile75+1.5*iqr
lower_limit=percentile25-1.5*iqr
print(upper_limit)
print(lower_limit)
```

```python
def replace_outliers_with_median(column):
    column = column.apply(lambda x: np.median(column) if x < lower_limit or x > upper_limit else x)
    return column

numeric_columns = df.select_dtypes(include=[np.number])
df[numeric_columns.columns] = numeric_columns.apply(replace_outliers_with_median)

print(df)
```

```python
df=df.drop(['VIP','Cabin'],axis=1)
df.head(10)
```

```python
df.Transported.value_counts()
```

```python
x=df.drop(columns=['Transported'],axis=1)
y=df.Transported
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
test_pred=model.predict(x_test)
test_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
cm=confusion_matrix(y_true=y_test,y_pred=test_pred) 
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=[False,True]) 
cm_display.plot() 
plt.show()
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(test_pred,y_test)
print("Accuracy percentage= ",score*100,"%")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
tree_classifier=DecisionTreeClassifier(max_leaf_nodes=35,random_state=23)
tree_classifier.fit(x_train,y_train)
```

```python
y_pred=tree_classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy score={accuracy_score(y_test,y_pred)*100}%")
```

```python
tree.plot_tree(tree_classifier)
plt.show()
```
