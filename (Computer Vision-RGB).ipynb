```python
plt.imshow(img_array)
```

```python
#Creating duplicates of the same image to individually remove R,G and B from each
duplicate_img1=img_array.copy()
duplicate_img2=img_array.copy()
duplicate_img3=img_array.copy()
```

```python
duplicate_img1[:,:,0]
```

```python
duplicate_img1[:,:,1]
```

```python
#Removing G and B filters to only show the R filter
duplicate_img1[:,:,1]=0
duplicate_img1[:,:,2]=0
plt.imshow(duplicate_img1)
```

```python
#Removing R and B filters to only show the G filter
duplicate_img2[:,:,0]=0
duplicate_img2[:,:,2]=0
plt.imshow(duplicate_img2)
```

```python
#Removing R and G filters to only show the B filter
duplicate_img3[:,:,0]=0
duplicate_img3[:,:,1]=0
plt.imshow(duplicate_img3)
```
