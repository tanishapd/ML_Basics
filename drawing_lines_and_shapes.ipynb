```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1=np.zeros(shape=(500,500,3))
```

```python
img1.shape
```

```python
plt.imshow(img1)
```

```python
cv2.rectangle(img1,pt1=(200,50),pt2=(300,300),color=(0,0,255),thickness=5)
plt.imshow(img1)
```

```python
cv2.circle(img1,center=(400,400),radius=50,color=(255,255,0),thickness=3)
plt.imshow(img1)
```

```python
cv2.circle(img1,center=(400,400),radius=50,color=(255,255,0),thickness=-1)
plt.imshow(img1)
```

```python
cv2.line(img1,pt1=(100,300),pt2=(400,100),color=(0,100,50),thickness=4)
plt.imshow(img1)
```
