# Import the necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('cat_meme.jpg')
```

```python
plt.imshow(img)
```

```python
#By default,the channel of the picture is assumed as BGR instead of RGB
#So let us convert the BGR channel to RGB
#This give us the original picture
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img=cv2.imread('cat_meme.jpg',0)
plt.imshow(img)
```

```python
#Converting to black and white
img=cv2.imread('cat_meme.jpg',0)
plt.imshow(img,cmap='gray')
```

```python
#Here
#100 -> Threshold value
#255 -> Maximum value you want to assign
lim,img1=cv2.threshold(img,100,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(img1,cmap='gray')
```

```python
lim,img2=cv2.threshold(img,100,255,cv2.THRESH_BINARY_INV)
plt.imshow(img2,cmap='gray')
```
